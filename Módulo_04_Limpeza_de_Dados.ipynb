## Exemplo 1

O primeiro exemplo que temos são dados que estão disponíveis dentro do portal [Inside Airbnb](http://insideairbnb.com). O Inside Airbnb não tem conexão oficial ou formal com o Airbnb. É uma iniciativa que coleta dados da plataforma e diponibiliza para o público, a fim de fomentar o segmento, e incentivar a comunidade a gerar análises e insights.

<center><img width="45%" src="https://image.freepik.com/free-photo/aerial-photo-rio-de-janeiro-surrounded-by-sea-hills-sunlight-brazil_181624-11362.jpg"></center>

Para o nosso notebook, vamos utilizar dados que estão disponíveis no portal sobre a cidade do [Rio De Janeiro](http://data.insideairbnb.com/brazil/rj/rio-de-janeiro/2020-06-19/data/listings.csv.gz).


````
# importando o pandas
import pandas as pd

# para não limitar a visualização do número de colunas
pd.set_option('display.max_columns', None)
````

````
# carregando os dados
df = pd.read_csv("http://data.insideairbnb.com/brazil/rj/rio-de-janeiro/2020-06-19/data/listings.csv.gz")
````

````
# visualizando as primeiras entradas
df.head()
````

Visualizando as entradas já podemos ter ideias das primeiras coisas que precisaremos arrumar. Na coluna `price`, por exemplo, temos um $, o que indica que o formato da coluna é `String`, e precisamos transformar em formato numérico para que possamos tirar vantagem dessa coluna.


````
# verificando as entradas da coluna price
df["price"]
````

Entretanto, não podemos fazer isso diretamente, justamente por causa desse sinal. Por isso, precisamos manipular e limpar essa string, para aí sim transformar em float.

````
# tentando converter para float
float("$244.00")
````

Para entendermos o processo que precisaremos fazer, vamos simplificar. Vamos criar uma variável e associar a ela uma string similar ao formato que encontramos no nosso data frame.

````
# associando o valor à variável
valor = "$244.00"
````

Agora, vamos utilizar o método `replace()` para substituir o cifrão por um espaço vazio, efetivamente removendo o cifrão da string.

Sem ele, podemos transformar essa string em float sem problemas.

````
# limpando a variável
valor = float(valor.replace("$", ""))
print(valor)
````

### Funções Lambda

De forma breve, funções Lambda são funções anônimas, que estão sujeitas a uma sintaxe mais restritiva, mas mais concisa do que as funções padrão do Python.

Quando vamos precisar reutilizar a função várias vezes no código, é preferível que façamos a construção de uma função padrão, mas para um objetivo mais direto, como é o caso de aplicarmos uma função específica apenas para fazer uma transformação ou limpeza em uma coluna, por exemplo, é bastante interessante se aproveitar do poder e simplicidade das funções lambda.

Para aprender mais sobre esse tipo de função, acesse [esse link](https://realpython.com/python-lambda/). 

````
# "$244.00" ==> 244.00
def converter_str(minha_string):
    x = float(minha_string.replace("$", ""))
    return x
````

````
# criando função lambda 
lambda x: float(x.replace("$", ""))
````

Com a função pronta, vamos aplicar na coluna e verificar os resultados.

````
df["price"].apply(converter_str)
````

````
# aplicando a função à coluna
df["price"].apply(lambda x: float(x.replace("$", "")))
````

A função está funcionando, tudo está dentro dos conformes, mas aqui encontramos uma situação inesperada. Retornamos um erro ao tentar converter um valor que tem **mais de 3 dígitos** antes da vírgula, está acima da casa de milhares.

Isso acontece por causa da vírgula. Floats são divididos a partir do ponto `.`, e a vírgula retorna um erro.

Vamos tratar esse detalhe para continuar a limpeza.

````
# associando o novo valor à variável
valor = "$4,161.00"
````

````
# acrescentando um passo à função
float(valor.replace(",", "").replace("$", ""))
````

Agora sim nossa função será capaz de entregar os valores no formato necessário para trabalhar com floats.

````
# aplicando a função sobre a coluna
df["price"].apply(lambda x: float(x.replace(",", "").replace("$", "")))
````

````
# carregando os dados
df = pd.read_excel("https://www.dropbox.com/s/69pajozeme6ezjc/solicitacoes-de-reconhecimento-da-condicao-de-refugiado.xlsx?dl=1")
````

Se comparado ao primeiro exemplo, esse conjunto de dados é mais simples. Aqui também vamos fazer manipulação de strings.

Vamos começar deixando apenas a primeira lentra da coluna `Tipo de Alertas e Rrestrições` em maiúsculo.

````
# verificando as entradas
df["Tipo de Alertas e Restrições"]
````

Para realizar essa operação, vamos unir nosso conhecimento de manipulação de strings, utilizando o método `capitalize` dentro de uma função lambda.

````
# aplicando capitalize 
df["Tipo de Alertas e Restrições"].apply(lambda x: x.capitalize())
````

É possível ver através do outuput que função retornou o resultado que queríamos. Entretanto, ainda não fizemos alteração alguma no dataset original.

É preciso associar à coluna do data frame o resultado da nossa operação sobre a mesma. Fazemos isso da seguinte forma:



```
df["coluna"] = df["coluna"].apply(lambda x: x.capitalize())
```

Dessa forma, associaremos à coluna os valores obtidos como output da função.

````
# aplicando a função e associando à coluna
df["Tipo de Alertas e Restrições"] = df["Tipo de Alertas e Restrições"].apply(lambda x: x.capitalize())
````

````
# verificando o resultado
df.head()
````

Agora sim podemos ver que o data frame foi alterado, e obtivemos o resultado esperado.

Agora vamos fazer o mesmo com a coluna `Nacionalidade`.

````
# aplicando a função
df["Nacionalidade"] = df["Nacionalidade"].apply(lambda x: x.title())
````

Para nosso terceiro e último exemplo do notebook, utilizaremos dados relacionadas à taxa de homicídios no Brasil em 2017. Mais uma vez, os dados originais podem ser encontrados no [Kaggle](https://www.kaggle.com/diegomariano/taxa-de-homicdios-no-brasil-no-ano-de-2017).

Aqui, vamos nos deparar com um outro tipo de dificuldade, oriundo da origem dos dados. Precisamos estar atentos a isso também, pois podemos encontrar arquivos com diferentes tipos de encoding, separadores, símbolos, nomenclaturas, etc.

Mais uma vez, chamamos atenção para a importância de conhecer seus dados a fundo.

````
# importando os dados
df = pd.read_csv("https://www.dropbox.com/s/16nz0yikhma091u/homicidios_brasil_2017.csv?dl=1")
````

````
# verificando as primeiras entradas
df.head()
````

Apesar de termos utilizado o mesmo método de sempre para abertura do arquivo, aqui, ele veio mal organizado, em um formato que não conseguimos trabalhar com o Pandas.

Isso ocorre porque arquivos `csv`, como o nome sugere (*Comma Separated Values*), tem como separador padrão a vírgula `","`, e o arquivo que estamos manipulando está usando ponto e vírgula `"/"` como separador.

Esse é um problema bastante simples que pode ser resolvido de forma rápida e direta. Basta definir o separador do arquivo como "/"



```
df = pd.read_csv('url_do_arquivo', sep=";")
```



````
# carregando o arquivo com o separador correto
df = pd.read_csv("https://www.dropbox.com/s/16nz0yikhma091u/homicidios_brasil_2017.csv?dl=1", sep=";")
````

````
# verificando o data frame
df
````

Acima, ao fazer uma verificação rápida no conjunto de dados conseguimos percerber que a sigla para os estados de Goiás (G) e São Paulo não estão corretamente lançadas na coluna `Sigla`. 

Precisamos consertar esse detalhe para que possamos fazer uma análise bem estruturada desses dados.

Para isso, vamos criar um dicionário que receberá como chave a entrada incorreta, e como valor, a correção.

Depois, utilizaremos o método `replace` para aplicar a correção.


````
# criando o dicionário para correção
correcao = {"G": "GO", "São Paulo": "SP"}

# aplicando a correção
df["sigla"] = df["sigla"].replace(correcao)
````

````
# verificando o resultado
df
````
