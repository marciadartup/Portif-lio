Na sequência, vamos importar o `pandas_profiling` e o conjunto de dados do CENIPA.

```
# importar as bibliotecas necessárias
import pandas as pd
from pandas_profiling import ProfileReport
from google.colab import drive

# importar dataset
#df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv")

# montando o drive
drive.mount('/content/gdrive')

df = pd.read_csv("/content/gdrive/My Drive/vendasexato.csv")
```

A primeira etapa, é criar um objeto do tipo ProfileReport e passar o próprio DataFrame.


```
df.head()
```

```
# criar ProfileReport
report = ProfileReport(df)
```

Para que os resultados sejam apresentados dentro do próprio notebook, basta executar o método `to_notebook_iframe()`.

```
# criar visualização iframe
report.to_notebook_iframe()
```

Caso você queira exportar o relatório para o formato `.html`, basta executar o método `to_file("nome.html")`

```
# criar relatório html
report.to_file("report.html")
```
